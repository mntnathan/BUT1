## 1.1. Manipulation arithmétique

Un ordinateur est simplement une machine très organisée qui obéit aux instructions et range ses données à sa manière. Cette machine ne comprend que le language *binaire*, c'est-à-dire des suites de $0$ et de $1$. Les unités élémentaires d'espace mémoire sont appelés *bits* et désigne un espace mémoire où seulement $0$ ou $1$ peuvent être stockés.

Les objets que nous connaisons comme les caractères a,b,...,z, (,),",' sont codés généralement sur $8$ bits que l'on appelle *octet*, voir les tables ASCII pour plus de détails.

Les nombres entiers sont quant à eux écrits en base $2$ (contrairement à notre système en base $10$). Par exemple, usuellement l'entier $12$ s'écrit dans notre système comme:
$$12=1\times 10+2\times 1.$$
L'ordinateur quant à lui verra l'entier $12$ comme:
$$12=1\times 2^3 + 1\times 2^2 +0 \times 2. $$
L'ordinateur stockera alors en mémoire $12$ comme $110$ mais en utilisant son language de $0$ et $1$.

Les nombres réels sont écrits sous forme de flottants, une approximation utilisant la base $2$ (le nombre d'octets dépend de la précision utilisée).

Un ordinateur sait manipuler ces données comme des entiers en faisant des additions, multiplications, divisions...

# 2. Boucles et tests conditionnels

Comme dans tout langage de programmation, des instructions sont fournies pour les tests conditionnels et les boucles:
- *if* est l'instruction de test conditionnel de base, équivalente du *si* en pseudocode;
- *else* est un instruction de test conditionnel pouvant être utilisée avec *if*. C'est un équivalent à *sinon* en pseudocode;
- *elif* est un instruction de test conditionnel pouvant être utilisée pour éviter d'imbriquer les conditions;
- *while* est une instruction qui boucle jusqu'à ce que la condition soit satisfaite, équivalente du *tant que* en pseudocode;
- *for* est une instruction qui boucle un nombre de fois donné, équivalente du *pour* en pseudocode.

Profitons-en pour remarquer les choses suivantes:
- si un programme boucle sans avoir l'air de progresser, c'est qu'une fonction *while* n'atteint pas sa condition d'arrêt;
- pour effectuer un test d'égalité OU d'inégalité, on utilise la syntaxe *==*, *!=*, *<*, *>*, *<=* et *=>*.

A partir de ces boucles algorithmiques, nous pouvons programmer des fonctions plus complexes.

Les instructions $if$ et $while$ sont un peu particulières puisque que ces dernières prennent en entrée un booléen en argument. Si ce booléen est *True*, elles effectuent la portion de code dans leur bloc, si le booléen est *False*, elles ignorent le bloc d'instructions qu'elle contiennent. Un bloc d'instructions est contenu dans un de ces tests si le bloc est indenté juste après le test en question.

La syntaxe de ces instructions impose de mettre des *:* et de repecter l'indentation des blocs. Voir les exemples pour plus de détails.

## 2.4. Mélange de toutes ces boucles:

Nous pouvons ensuite mélanger toutes ces différentes instructions afin de faire des programmes plus compliqués.

Dire ce que fait le code suivant (sans l'exécuter):

```python
for i in range(20):
    j=0
    if i%2==0:
        print(i)
    else:
        j=j+i
print(j)
# Ce programme renvoie tout les nombres pair jusqu'a 20 et apres il il affiche le nombre impair
```

Ce programme peut également être écrit en ayant recourt aux listes:

```python
l=[]
for i in range(20):
    j=0
    if i%2==0:
        l.append(i) # .append(i) ajoute la valeur i à la fin de la liste
    else:
        j=j+i
print(j)
print(l)
```
